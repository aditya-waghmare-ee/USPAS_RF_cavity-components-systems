Calculations for Part 4 <br> <br>
The cut off-wave number is $\ k=\omega  \sqrt{\mu \epsilon} \ $ <br>
The cut off frequency is the frequency of cut off of the $\ TE_{{10}} \ $ mode which is given by <br>
$\ f_{{cutoff}} = \frac {1} {2a \sqrt{{\mu \epsilon}}} \ $

```python
import numpy as np
from IPython.display import display, Math

# CutOff Frequency Calculation using physical waveguide dimension 'a'
mu_0=4*np.pi*1e-7
epsilon_0=8.854e-12

wr2300_a=584.2e-3
calc_fco=(0.5/((wr2300_a)*np.sqrt(mu_0*epsilon_0)))
display(Math(rf"f_{{\mathrm{{cutoff}}}} = {calc_fco*1e-9}\ \ GHz"))

freq_p=352*1e+6
display(Math(rf"f_{{\mathrm{{propogration}}}} = {freq_p*1e-9}\ \ GHz"))

rad_fp = 2 * np.pi * freq_p
wavenum=rad_fp*np.sqrt(mu_0*epsilon_0)
display(Math(rf"k = {wavenum}"))

#CST results
CST_fco=0.256449e+9
CST_wavenum=(2*np.pi*freq_p)*np.sqrt(mu_0*epsilon_0)
CST_beta=4.9917
CST_k_from_beta=np.sqrt((CST_beta**2)+(((np.pi)**2)/(wr2300_a**2)))
display(Math(rf"CST \ f_{{\mathrm{{cutoff}}}} = {CST_fco*1e-9}\ \ GHz"))
display(Math(rf"CST \ k_{{propogating}} = {CST_k_from_beta}"))

```

We can see that our values calculated in part 4 match very well with the CST results obtained, verifying simulation results.
