#### Problem 2.21
$$
\begin{array}{l}
Z_{in} = Z_0 \frac{Z_L + j Z_0 \tan(\beta l)}{Z_0 + j Z_L \tan(\beta l)} && (1)\\ \\
but \quad Z_L \rightarrow 0 \hspace{10em} && (given) \\
\tan(\beta l) = \frac{Z_{in}}{j Z_0} \\ 
\beta l =tan^{-1}(\frac {Z_{in}} {j Z_0}) && (2)
\end{array}
$$
